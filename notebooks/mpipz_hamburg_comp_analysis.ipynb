```python
from pathlib import Path

import pandas as pd
import seaborn as sns
```

```python
df_mpipz = pd.read_feather("../data/mpipz/filtered-joined-counts/df1.feather")
df_mpipz.shape
```

```python
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

df_mpipz_corr_filtered = df_mpipz

# Convert df_mpipz_corr to colors
df_mpipz_corr_colors = pd.DataFrame(index=df_mpipz_corr_filtered.index)

rd_bu_cmap = sns.color_palette("RdBu", as_cmap=True)
norm = Normalize(vmin=-1, vmax=1)
scalar_mappable = ScalarMappable(norm=norm, cmap=rd_bu_cmap)

for col in sorted(df_mpipz_corr_filtered.columns):
    df_mpipz_corr_colors[col] = df_mpipz_corr_filtered[col].map(scalar_mappable.to_rgba)

print(df_mpipz_corr_colors.shape)
df_mpipz_corr_colors.head()
```

```python
df_diff_hamburg_c_cl_counts = pd.read_feather("../data/hamburg/Col_ColLer.diff_counts.feather")
print(df_diff_hamburg_c_cl_counts.shape)
df_diff_hamburg_cl_c_counts = pd.read_feather("../data/hamburg/ColLer_Col.diff_counts.feather")
print(df_diff_hamburg_cl_c_counts.shape)

df_diff_hamburg_c_cl_corrs = pd.read_feather("../data/hamburg/Col_ColLer.diff_corrs.feather")
print(df_diff_hamburg_c_cl_corrs.shape)
df_diff_hamburg_cl_c_corrs = pd.read_feather("../data/hamburg/ColLer_Col.diff_corrs.feather")
print(df_diff_hamburg_cl_c_corrs.shape)

df_diff_mpipz_female_counts = pd.read_feather("../data/mpipz/wt_female_plate_3_7.diff_counts.feather")
print(df_diff_mpipz_female_counts.shape)
df_diff_mpipz_male_counts = pd.read_feather("../data/mpipz/wt_male_plate_3_7.diff_counts.feather")
print(df_diff_mpipz_male_counts.shape)

df_diff_mpipz_female_corrs = pd.read_feather("../data/mpipz/wt_female_plate_3_7.diff_corrs.feather")
print(df_diff_mpipz_female_corrs.shape)
df_diff_mpipz_male_corrs = pd.read_feather("../data/mpipz/wt_male_plate_3_7.diff_corrs.feather")
print(df_diff_mpipz_male_corrs.shape)
```

```python
diff_intersection = df_diff_hamburg_c_cl_counts.index.intersection(
    df_diff_hamburg_c_cl_corrs.index
).intersection(
    df_diff_hamburg_cl_c_counts.index
).intersection(
    df_diff_hamburg_cl_c_corrs.index
).intersection(
    df_diff_mpipz_female_counts.index
).intersection(
    df_diff_mpipz_male_counts.index
).intersection(
    df_diff_mpipz_female_corrs.index
).intersection(
    df_diff_mpipz_male_corrs.index
)

print(diff_intersection.shape)
```

```python
def count_stats_of(df: pd.DataFrame, prefix: str) -> pd.DataFrame:
    return pd.DataFrame(
        {
            "mean": df.mean(axis=1),
            "std": df.std(axis=1),
            "min": df.min(axis=1),
            "max": df.max(axis=1),
            "n": df.count(axis=1),
        }
    ).add_prefix(prefix)
```

```python
df_diff_summary = count_stats_of(df_diff_hamburg_cl_c_counts.fillna(0), "hamburg_cl_c_").join(
    count_stats_of(df_diff_hamburg_c_cl_counts.fillna(0), "hamburg_c_cl_"), how="inner"
).join(
    count_stats_of(df_diff_mpipz_female_counts.fillna(0), "mpipz_female_"), how="inner"
).join(
    count_stats_of(df_diff_mpipz_male_counts.fillna(0), "mpipz_male_"), how="inner"
).join(
    df_diff_hamburg_c_cl_corrs.add_prefix("hamburg_c_cl_corr_"), how="inner"
).join(
    df_diff_hamburg_cl_c_corrs.add_prefix("hamburg_cl_c_corr_"), how="inner"
).join(
    df_diff_mpipz_female_corrs.add_prefix("mpipz_female_corr_"), how="inner"
).join(
    df_diff_mpipz_male_corrs.add_prefix("mpipz_male_corr_"), how="inner"
)

df_diff_summary
```
