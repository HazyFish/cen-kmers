### Get Differential K-mers

```python
df_diff: Optional[pd.DataFrame] = None

for path in Path("../data/7-eco/joined-counts/").glob("*.part.*.feather"):
    print(path)
    part = pd.read_feather(path)
    print(part.shape)
    part.fillna(0, inplace=True)
    part = part[(part == 0).sum(axis=1) > 0]
    print(part.shape)

    if df_diff is None:
        df_diff = part
    else:
        df_diff = pd.concat([df_diff, part])

assert df_diff is not None

df_diff.shape
```

```python
df_diff.head()
```

```python
df_diff.to_feather(Path("../data/7-eco/ecotypes-joined-counts-diff.feather"), compression="zstd", compression_level=8)
```

### Join Kmers with Hamburg

```python
df_hamburg = pd.read_feather(Path("../data/hamburg/correlations/inner-joined-corr.feather"))
df_hamburg.shape
```

```python
df_hamburg.head()
```

```python
df_joined: Optional[pd.DataFrame] = None

for path in Path("../data/7-eco/joined-counts/").glob("*.part.*.feather"):
    print(path)
    part = pd.read_feather(path)
    print(part.shape)
    part.fillna(0, inplace=True)
    part = part[part.index.isin(df_hamburg.index)]
    print(part.shape)

    if df_joined is None:
        df_joined = part
    else:
        df_joined = pd.concat([df_joined, part])

assert df_joined is not None

df_joined.shape
```

```python
df_joined.to_feather(Path("../data/7-eco/ecotypes-joined-counts-hamburg.feather"), compression="zstd", compression_level=8)
```
