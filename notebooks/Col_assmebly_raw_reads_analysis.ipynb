```python
import pandas as pd
```

```python
df_colpek_reads = pd.read_feather("../SRR16832054.ge10.feather")
df_colpek_reads.shape
```

```python
df_mutac3 = pd.read_feather("../data/joined-counts.df3.zstd8.feather")
df_mutac3.shape
```

```python
df_joined = df_mutac3.join(df_colpek_reads, how="inner")
df_joined.shape
```

```python
df_joined.head()
```

```python
df_joined.to_feather("../data/colpek_mutac3_joined.feather", compression="zstd", compression_level=8)
```

```python
df_joined.to_csv("../data/colpek_mutac3_joined.csv")
```
