```python
from pathlib import Path

import pandas as pd
import seaborn as sns
```

```python
df_mpipz_filtered_corr = pd.read_feather("../data/mpipz/filtered-joined-counts/df2.feather")
df_mpipz_filtered_corr
```

```python
df_hamburg_filtered_corr = pd.read_feather(Path("../data/hamburg/correlations/df12.feather"))
df_hamburg_filtered_corr
```

```python
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

df_hamburg_corr_colors = pd.DataFrame(index=df_hamburg_filtered_corr.index)

rd_bu_cmap = sns.color_palette("RdBu", as_cmap=True)
norm = Normalize(vmin=-1, vmax=1)
scalar_mappable = ScalarMappable(norm=norm, cmap=rd_bu_cmap)

for col in sorted(df_hamburg_filtered_corr.columns):
    df_hamburg_corr_colors[col] = df_hamburg_filtered_corr[col].map(scalar_mappable.to_rgba)

print(df_hamburg_corr_colors.shape)
df_hamburg_corr_colors.head()
```

```python
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

df_mpipz_corr_colors = pd.DataFrame(index=df_mpipz_filtered_corr.index)

rd_bu_cmap = sns.color_palette("RdBu", as_cmap=True)
norm = Normalize(vmin=-1, vmax=1)
scalar_mappable = ScalarMappable(norm=norm, cmap=rd_bu_cmap)

for col in sorted(df_mpipz_filtered_corr.columns):
    df_mpipz_corr_colors[col] = df_mpipz_filtered_corr[col].map(scalar_mappable.to_rgba)

print(df_mpipz_corr_colors.shape)
df_mpipz_corr_colors.head()
```

```python
df_joined = df_hamburg_filtered_corr.join(df_mpipz_filtered_corr, how="inner")
df_joined.shape
```

```python
sns.clustermap(
    df_joined.reindex(sorted(df_joined.columns), axis=1),
    col_cluster=False,
    cmap="RdBu",
    figsize=(16, 24),
    vmin=-1,
    vmax=1,
)
```

```python
df_discordant = df_joined[df_joined.max(axis=1) - df_joined.min(axis=1) > 1]
df_discordant = df_discordant.rename(columns=lambda col: col.replace("_", "M_") if col.startswith("Chr") else f"Chr{col}")
df_discordant.shape
```

```python
df_discordant.sample(10)
```

```python
sns.clustermap(
    df_discordant.reindex(sorted(df_discordant.columns), axis=1),
    col_cluster=False,
    cmap="RdBu",
    figsize=(16, 24),
    vmin=-1,
    vmax=1,
)
```

```python
chr_dfs = {
    i: df_joined.filter(like=str(i)) for i in range(1, 6)
}

chr_range_dfs = {
    i: df_chr.max(axis=1) - df_chr.min(axis=1) for i, df_chr in chr_dfs.items()
}

df_chr_ranges = pd.DataFrame(chr_range_dfs)
df_chr_ranges.sample(10)
```

```python
df_d2 = df_joined[
    df_chr_ranges.max(axis=1) > 0.5
]
df_d2 = df_d2.rename(columns=lambda col: col.replace("_", "M_") if col.startswith("Chr") else f"Chr{col}")
df_d2.shape
```

```python
sns.clustermap(
    df_d2.reindex(sorted(df_d2.columns), axis=1),
    col_cluster=False,
    cmap="RdBu",
    figsize=(16, 24),
    vmin=-1,
    vmax=1,
)
```
